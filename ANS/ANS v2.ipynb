```python
# Functions

#
# Function that takes user info
#

save_data = False
user_data = {}

def user_info():
    """ Takes user input and adds data to user_data dictionary

    param: no parameters
    """
    global save_data
    display(HTML("<span>Enter a username:</span>"))
    username = input()
    
    user_data["participant id"] = username
    display(HTML("<span>Enter your gender: male/female/other </span>"))
    gender = input()
    user_data["gender"] = gender.lower()
    
    display(HTML("<span>Enter your age:</span."))
    age = input()
    user_data["age"] = age

    display(HTML("<span>How much sleep have you had in hours:</span>"))
    sleep_in_hours = input()
    user_data["sleep in hours"] = sleep_in_hours

    display(HTML("<span>Have you had a caffeinated beverages today: Y/N"))
    caffeinated = input().upper()
    user_data["caffeinated"] = caffeinated
                     
    display(HTML("<span>Please read:</span."))
    print("")
    display(HTML("<span>We wish to record your response data to an anonymised public data repository.</span>"))
    print("")
    display(HTML("<span>Your data will be used for educational teaching purposes practising data analysis and visualisation.</span>"))
    print("")
    display(HTML("<span>Please type YES in the box below if you consent to the upload and NO if you do not consent.</span>"))
    
    result = input("> ").upper()
    if result == "YES":
        display(HTML("Thanks - your data will be uploaded."))
        save_data = True
        clear_output(wait=True)
    else:
        display(HTML("No problem we hope you enjoyed the test."))
        clear_output(wait=True)
    return



def countdown(length_in_seconds):
    """ Runs a countdown that rewrites itself by moving the carriage return

    param: desired length of countdown in seconds
    """
    for i in range(length_in_seconds):
        print(length_in_seconds," ", end='\r')
        time.sleep(1)
        length_in_seconds -= 1  
    return


def Introduction():
    """ Displays the introduction for the test,

    param: no parameters
    """
    display(HTML("<h1>Welcome to the Approximate Number System (ANS) test</h1>"))
    time.sleep(5)
    display(HTML("<span>In this test you will be shown two circles with dots inside."))
    display(HTML("<span>You must indicate which circle has the most dots by clicking on the 'Left' and 'Right' buttons below the image</span>"))
    display(HTML("<span>You will be shown each image for 0.75 seconds, 3 seconds to give your answer, and a 1.5 second break between images</span>"))
    time.sleep(8)

#
# Function that checks the contents of event_info every 0.001 seconds to check for input via the buttons
#

event_info = {
    'type': '',
    'description': '',
    'time': -1
}

def wait_for_event(timeout=-1, interval=0.001, max_rate=20, allow_interupt=True):  
    """ Checks the contents of event_info every 0.001 seconds,
    to check for input via the buttons
    """
    start_wait = time.time()
    event_info['type'] = ""
    event_info['description'] = ""
    event_info['time'] = -1

    n_proc = int(max_rate*interval)+1
    
    with ui_events() as ui_poll:
        keep_looping = True
        while keep_looping==True:
            # process UI events
            ui_poll(n_proc)

            # end loop if we have waited more than the timeout period
            if (timeout != -1) and (time.time() > start_wait + timeout):
                keep_looping = False
                
            # end loop if event has occured
            if allow_interupt==True and event_info['description']!="":
                keep_looping = False
                
            # add pause before looping
            # to check events again
            time.sleep(interval)
    
    # return event description after wait ends
    # will be set to empty string '' if no event occured
    return event_info



#
# Set up buttons
#

btnL = widgets.Button(description="L", layout=Layout(width="500px", height='80px'))
btnR = widgets.Button(description="R", layout=Layout(width="500px", height='80px'))



#
# function lets buttons register events when clicked
#

def register_event(btn):
    """ Registers when button has been clicked

    param: button click
    """
    event_info['type'] = "click"
    event_info['description'] = btn.description
    event_info['time'] = time.time()
    return

#  
# button to calls the register_event function when clicked
#

btnL.on_click(register_event) 
btnR.on_click(register_event) 

#
# function to display the image argument and blank cross images as well as display the buttons and call wait_for_event() for timed input
#

def test(image,question_number):
    """ Displays the image argument and blank cross images,
    displays the buttons and calls the wait_for_event() for timed input

    param: test image (image with blue and yellow dots)
    """
    panel = widgets.HBox([btnL, btnR])
    
    display(HTML(f"<h1>Q{question_number}</h1>")) # Added in response to feedback
    display(image)
    
    start_time = time.time()
    
    time.sleep(0.75)
    clear_output(wait=True)
    
    display(HTML("<h1>---</h1>"))
    display(blank_cross)
    
    panel = widgets.HBox([btnL, btnR])
    display(panel)
    answer = wait_for_event(timeout=3)
    
    finish_time = time.time()
    clear_output(wait=True)
    
    display(HTML("<h1>BREAK</h1>"))
    display(blank_cross)
    time.sleep(1.5)
    clear_output(wait=True)
    
    time_on_q = finish_time - start_time
    
    return time_on_q, answer["description"]
      
#  
# Function to run through all the images in a dictionary containing the images and answers by calling test()
#

def run_test(questions):
    """ Runs the test by calling  Introduction(), user_info(), and test(),

    param: list of the order of the images shown
    """
    Introduction()
    user_info() 
    
    display(HTML("<span>Make sure to scroll down before the test starts so that you can see the countdown timer</span>"))
    display(blank_cross)
    display(HTML("<h1>Test Begins in 10 seconds</h1>"))
    countdown(10)
    clear_output(wait=True)
    
    for x, i in enumerate(questions):
        q_no = x + 1
        time, answers  = test(i,q_no)
        rounded_time = round(time, 3)
        user_times.append(rounded_time)
        user_answers.append(answers)
    return
#
# Function that takes user_answers and user_times lists to create a list with the users incorrect, correct, 
# and timeout answers as well as calculate the users score
#

def data_processing(user_answers, user_times):
    """ Processes the data collected from the test,
    marks the answers from the user,
    calculates percentage score,
    adds the results and the question identifiers to the user_data dicationary
    
    param: list of the users answers and list of the users times
    """
    marked_ans = []
    count = 0
    for i in range(len(user_times)):
        if user_answers[i] == answer_key[i]:
            marked_ans.append("correct")
            count += 1
        elif user_answers[i] == "":
            marked_ans.append("timeout")
        else:
            marked_ans.append("incorrect")
    
    
    
     
    for i in range(len(user_times)):
        user_data[f"q{i+1}_image"] = image_dot_no[i]
        user_data[f"q{i+1}_ans"] = user_answers[i]
        user_data[f"q{i+1}_correct_ans"] = answer_key[i]
        user_data[f"q{i+1}_result"] = marked_ans[i]
        user_data[f"q{i+1}_time"] = user_times[i]

    percentage_score = (count/32) * 100 # added in response to feedback
    return percentage_score



form_url = "https://docs.google.com/forms/d/e/1FAIpQLSf33bG1HFaP7k6KDWG1fAoIPebkYX3RG66XTFvRw-k4_isuNg/viewform"

def send_to_google_form(data_dict, form_url):
    """ sends data from dictinoary to google form, returns whether upload was successful

    param: dictionary of user results and google form url
    """
   
    form_id = form_url.split('/e/')[1].split('/')[0]
    view_form_url = f'https://docs.google.com/forms/d/e/{form_id}/viewform'
    post_form_url = f'https://docs.google.com/forms/d/e/{form_id}/formResponse'

    page = requests.get(view_form_url)
    content = BeautifulSoup(page.content, "html.parser").find('script', type='text/javascript')
    content = content.text[27:-1]
    result = json.loads(content)[1][1]
    form_dict = {}
    
    loaded_all = True
    for item in result:
        if item[1] not in data_dict:
            print(f"Form item {item[1]} not found. Data not uploaded.")
            loaded_all = False
            return False
        form_dict[f'entry.{item[4][0][0]}'] = data_dict[item[1]]
    
    post_result = requests.post(post_form_url, data=form_dict)
    return post_result.ok       
            
         
            
#
# Function to run the whole test and upload data
#
        
def ANS_test():
    """ runs the whole test and tells user percentage score 
    and whether their data was uploaded

    param: none
    """
    run_test(image_order)
    percentage_score = data_processing(user_answers, user_times)
    
    if save_data:
        send_to_google_form(user_data, form_url)
        display(HTML(f"<span>Your results have been uploaded, thankyou for completing the test</span>"))
        display(HTML(f"<span>You got {percentage_score}% on this test")) # Added in response to feedback
    else:
        display(HTML(f"<span>Thankyou for completing the test</span>"))
        display(HTML(f"<span>You got {percentage_score}% on this test"))
```
