```python
!pip install jupyter_ui_poll

from IPython.display import display, Image, clear_output, HTML
from ipywidgets import Layout
import time
import ipywidgets as widgets
from jupyter_ui_poll import ui_events
import requests
from bs4 import BeautifulSoup
import json 

clear_output(wait=True)


blank_cross = Image("Blank_with_Cross.png", width=1000)


width = 1000

i_10_9_1 = Image("10_to_9_1.png", width=width)
i_10_9_2 = Image("10_to_9_2.png", width=width)

i_12_14_1 = Image("12_to_14_1.png", width=width) 
i_12_14_2 = Image("12_to_14_2.png", width=width) 

i_12_16_1 = Image("12_to_16_1.png", width=width) 
i_12_16_2 = Image("12_to_16_2.png", width=width) 

i_12_9_1 = Image("12_to_9_1.png", width=width) 
i_12_9_2 = Image("12_to_9_2.png", width=width) 

i_14_12_1 = Image("14_to_12_1.png", width=width) 
i_14_12_2 = Image("14_to_12_2.png", width=width) 

i_15_20_1 = Image("15_to_20_1.png", width=width) 
i_15_20_2 = Image("15_to_20_2.png", width=width) 

i_16_12_1 = Image("16_to_12_1.png", width=width) 
i_16_12_2 = Image("16_to_12_2.png", width=width) 

i_16_18_1 = Image("16_to_18_1.png", width=width) 
i_16_18_2 = Image("16_to_18_2.png", width=width) 

i_18_16_1 = Image("18_to_16_1.png", width=width) 
i_18_16_2 = Image("18_to_16_2.png", width=width) 

i_18_20_1 = Image("18_to_20_1.png", width=width) 
i_18_20_2 = Image("18_to_20_2.png", width=width)

i_18_21_1 = Image("18_to_21_1.png", width=width) 
i_18_21_2 = Image("18_to_21_2.png", width=width) 

i_20_15_1 = Image("20_to_15_1.png", width=width) 
i_20_15_2 = Image("20_to_15_2.png", width=width) 

i_20_18_1 = Image("20_to_18_1.png", width=width) 
i_20_18_2 = Image("20_to_18_2.png", width=width) 

i_21_18_1 = Image("21_to_18_1.png", width=width) 
i_21_18_2 = Image("21_to_18_2.png", width=width) 

i_9_10_1 = Image("9_to_10_1.png", width=width) 
i_9_10_2 = Image("9_to_10_2.png", width=width) 

i_9_12_1 = Image("9_to_12_1.png", width=width) 
i_9_12_2 = Image("9_to_12_2.png", width=width) 


#
# Dictionaries for answer key and the users answers
#
answer_key = ["L", "L", "L", "R", "L", "R", "R", "R", "L", "R", "L", "R", "L", "R", "R", "L", "R", "R", "L", "L", "L",
              "R", "R", "L", "R", "L", "R", "R", "L", "L", "R", "L", "R", "L", "L", "R", "L", "L", "R", "R", "L"]

image_order = [i_16_12_2, i_18_16_1, i_18_16_2, i_16_18_1, i_20_18_1, i_15_20_1, i_9_12_2, i_12_14_2, i_20_15_1, i_18_20_2, i_12_16_2,
               i_20_15_2, i_12_9_2, i_18_21_2, i_20_18_2, i_12_16_1, i_12_14_1, i_16_18_2, i_9_12_1, i_21_18_2, i_10_9_1, i_14_12_1, i_15_20_2,
               i_9_10_2, i_10_9_2, i_18_21_1, i_12_9_1, i_9_10_1, i_21_18_1, i_14_12_2, i_18_20_1, i_16_12_1]

image_dot_no = ["i_16_12_2", "i_18_16_1", "i_18_16_2", "i_16_18_1", "i_20_18_1", "i_15_20_1", "i_9_12_2", "i_12_14_2", "i_20_15_1", "i_18_20_2", "i_12_16_2",
                "i_20_15_2", "i_12_9_2", "i_18_21_2", "i_20_18_2", "i_12_16_1", "i_12_14_1", "i_16_18_2", "i_9_12_1", "i_21_18_2", "i_10_9_1", "i_14_12_1", "i_15_20_2",
                "i_9_10_2", "i_10_9_2", "i_18_21_1", "i_12_9_1", "i_9_10_1", "i_21_18_1", "i_14_12_2", "i_18_20_1", "i_16_12_1"]

user_answers = []
user_times = []


#
# TO FINISH Function that takes user info
#

save_data = False
user_data = {}
def user_info():
    global save_data
    display(HTML("<span>Enter your name:</span."))
    username = input()
    user_data["participant id"] = username
    display(HTML("<span>Enter your gender: male/female </span."))
    gender = input()
    user_data["gender"] = gender.lower()
    display(HTML("<span>Enter your age:</span."))
    age = input()
    user_data["age"] = age
    display(HTML("<span>Please read:</span."))
    print("")
    display(HTML("<span>We wish to record your response data to an anonymised public data repository.</span."))
    print("")
    display(HTML("<span>Your data will be used for educational teaching purposes practising data analysis and visualisation.</span."))
    print("")
    display(HTML("<span>Please type YES in the box below if you consent to the upload and NO if you do not consent.</span."))
    
    result = input("> ")
    if result == "YES":
        display(HTML("Thanks - your data will be uploaded."))
        save_data = True
        clear_output(wait=True)
    else:
        display(HTML("No problem we hope you enjoyed the test."))
        clear_output(wait=True)
    return

def countdown(length_in_seconds):
    for i in range(length_in_seconds):
        print(length_in_seconds," ", end='\r')
        #end='\r' tells python to move the carriage back to the start of the line, this means that when it loops the line will be rewritten.
        #This makes the countdown cleaner becuase your not outputting 15 statemnts about the countdown.
        #the extra blacnk space after the length_in_seconds is there because when you go from 2 digit to 1 digit numbers, there will be a left over 0 from
        #the 10 that won't be rewritten without that extra space.
        time.sleep(1)
        length_in_seconds -= 1  
    return

def Introduction():
    display(HTML("<h1>Welcome to the Aproximate Number Sense (ANS) test</h1>"))
    time.sleep(5)
    display(HTML("<span>In this test you will be shown two circles with dots inside. <br>You must indicate which circle has the most dots<br> You will be shown each image for 0.75 seconds, 3 seconds to give your answer, and a 1.5 second break between images</span>"))
    time.sleep(8)




#
# Function that checks the contents of event_info every 0.001 seconds to check for input via the buttons
#
event_info = {
    'type': '',
    'description': '',
    'time': -1
}

def wait_for_event(timeout=-1, interval=0.001, max_rate=20, allow_interupt=True):    
    start_wait = time.time()

    # set event info to be empty
    # as this is dict we can change entries
    # directly without using
    # the global keyword
    event_info['type'] = ""
    event_info['description'] = ""
    event_info['time'] = -1

    n_proc = int(max_rate*interval)+1
    
    with ui_events() as ui_poll:
        keep_looping = True
        while keep_looping==True:
            # process UI events
            ui_poll(n_proc)

            # end loop if we have waited more than the timeout period
            if (timeout != -1) and (time.time() > start_wait + timeout):
                keep_looping = False
                
            # end loop if event has occured
            if allow_interupt==True and event_info['description']!="":
                keep_looping = False
                
            # add pause before looping
            # to check events again
            time.sleep(interval)
    
    # return event description after wait ends
    # will be set to empty string '' if no event occured
    return event_info

#
# function lets buttons register events when clicked
#
def register_event(btn):
    # display button description in output area
    event_info['type'] = "click"
    event_info['description'] = btn.description
    event_info['time'] = time.time()
    return


#
# Set up buttons
#
btnL = widgets.Button(description="L", layout=Layout(width="500px", height='80px'))
btnR = widgets.Button(description="R", layout=Layout(width="500px", height='80px'))

#  
# button to calls the register_event function when clicked
#
btnL.on_click(register_event) 
btnR.on_click(register_event) 

#
# function to display the test images and blank cross images as well as display the buttons and call wait_for_event() for timed input
#


def test(image):
    panel = widgets.HBox([btnL, btnR])
    
    display(HTML("<h1>---</h1>"))
    display(image)
    
    start_time = time.time()
    
    time.sleep(2)
    clear_output(wait=True)
    
    display(HTML("<h1>---</h1>"))
    display(blank_cross)
    
    panel = widgets.HBox([btnL, btnR])
    display(panel)
    answer = wait_for_event(timeout=3)
    
    finish_time = time.time()
    clear_output(wait=True)
    
    display(HTML("<h1>BREAK</h1>"))
    display(blank_cross)
    time.sleep(1.5)
    clear_output(wait=True)
    
    time_on_q = finish_time - start_time
    
    return time_on_q, answer["description"]
    
    
    
#  
# Function to run through all the images in a dictionary containing the images and answers
#
def run_test(questions):
    Introduction()
    user_info() 
    
    display(blank_cross)
    display(HTML("<h1>Test Begins in 10 seconds</h1>"))
    countdown(10)
    clear_output(wait=True)
    for i in questions:
        
        time, answers  = test(i)
        rounded_time = round(time, 3)
        user_times.append(rounded_time)
        user_answers.append(answers)

    print("thankyou for completing the test")
    if save_data == True:
        print("your results will be uploaded to the repository")

    if save_data == False:
        print("your data will not be uploaded to the repository as you did not consent to data collection")
    
    return


def data_processing(user_answers, user_times):
    marked_ans = []
    for i in range(len(user_times)):
        if user_answers[i] == answer_key[i]:
            marked_ans.append("correct")
        elif user_answers[i] == "":
            marked_ans.append("timeout")
        else:
            marked_ans.append("incorrect")
    
    for i in range(len(user_times)):
        user_data[f"q{i+1}_image"] = image_dot_no[i]
        user_data[f"q{i+1}_ans"] = user_answers[i]
        user_data[f"q{i+1}_correct_ans"] = answer_key[i]
        user_data[f"q{i+1}_result"] = marked_ans[i]
        user_data[f"q{i+1}_time"] = user_times[i]
    return

def send_to_google_form(data_dict, form_url):
   
    form_id = form_url.split('/e/')[1].split('/')[0]
    view_form_url = f'https://docs.google.com/forms/d/e/{form_id}/viewform'
    post_form_url = f'https://docs.google.com/forms/d/e/{form_id}/formResponse'

    page = requests.get(view_form_url)
    content = BeautifulSoup(page.content, "html.parser").find('script', type='text/javascript')
    content = content.text[27:-1]
    result = json.loads(content)[1][1]
    form_dict = {}
    
    loaded_all = True
    for item in result:
        if item[1] not in data_dict:
            print(f"Form item {item[1]} not found. Data not uploaded.")
            loaded_all = False
            return False
        form_dict[f'entry.{item[4][0][0]}'] = data_dict[item[1]]
    
    post_result = requests.post(post_form_url, data=form_dict)
    return post_result.ok       
            
         
            

        



run_test(image_order)
data_processing(user_answers, user_times)

print(user_answers)
print(user_times)
display(user_data)

form_url = "https://docs.google.com/forms/d/e/1FAIpQLSfZT-KMgVt02SqBVJHEkwot7Cb2rgWM4zjk0Vf4GZoH4oYiTg/viewform"


if save_data:
    send_to_google_form(user_data, form_url)
    display(HTML("<span>Your results have been uploaded, thankyou for completing the test</span>"))
else:
    display(HTML("<span>Thankyou for completing the test</span>"))


```
