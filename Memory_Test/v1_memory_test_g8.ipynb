```python
from IPython.display import display, Image, clear_output, HTML
import time

# Introductory and closing texts
message_0 = "Before we start, tell us about yourself."

message_1 = "Welcome to the Memory Test!"
message_2 = "You will be shown a grid containing 9 objects of different shapes and colors."
message_3 = "You will be given 10 seconds to memorize the image before being asked a series of memory-based questions."
message_4 = "Each question is alloted 2 attempts before being marked incorrect."
message_5 = "The test will begin in 5 seconds."
message_6 = "Start"

message_7 = "Congratulations!"
message_8 = "You have completed the Memory Test."

# Images
image1 = Image("image1.png", width = 300)
image1_q2 = Image("image1_q2.png", width = 300)
image1_q4 = Image("image1_q4.png", width = 300)

image2 = Image("image2.png", width = 300)
image2_q3 = Image("image2_q3.png", width = 300)
image2_q4 = Image("image2_q4.png", width = 300)

# Questions
question_1 = "How many shapes had the color green?"
question_2 = "What shape was in the indicated position?"
question_3 = "What color was the cross?"
question_4 = "What color was the shape in the indicated position?"
question_5 = "What shape had the color gray?"

question_6 = "How many colors were repeated?"
question_7 = "How many shapes had secondary shapes within them?"
question_8 = "What shape was in the indicated position?"
question_9 = "What color was the shape in the indicated position?"
question_10 = "What shape was the color pink?"

# Answers
answer_1 = "2"
answer_2 = "circle"
answer_3 = "pink"
answer_4 = "purple"
answer_5 = "rectangle"

answer_6 = "2"
answer_7 = "3"
answer_8 = "triangle"
answer_9 = "yellow"
answer_10 = "triangle"

# Storing all questions in a list
question_list = [question_1, question_2, question_3, question_4, question_5, 
                 question_6, question_7, question_8, question_9, question_10]

# Storing all answers in a list
answer_list = [answer_1, answer_2, answer_3, answer_4, answer_5, 
               answer_6, answer_7, answer_8, answer_9, answer_10]

# Empty lists for data storage
saved_names = []
saved_genders = []
saved_ages = []
saved_sleep = []
saved_caffeine = []
saved_scores = []

# Data dictionary for data storage
saved_data = {
    'Name' : saved_names,
    'Gender' : saved_genders,
    'Age' : saved_ages,
    'Sleep' : saved_sleep,
    'Caffeine' : saved_caffeine,
    'Score' : saved_scores
}
```

```python
# Function prints introduction to test 
# and collects data regarding user

def print_introduction():

    username = ""
    gender = ""
    age = ""
    sleep = ""
    caffeine = ""
    
    # Stores user's name as the variable 'username'
    username = input("Please enter your name: ")
    clear_output(wait=False)
    print(f"Hello {username}.")
    time.sleep(1.5)

    print(message_0)
    time.sleep(1.5)

    # Stores user's gender as the variable 'gender'
    print("\nWhat is your gender? Please enter your response using the corresponding number:")
    print("\t1 - Male")
    print("\t2 - Female")
    print("\t3 - Nonbinary")
    print("\t4 - Other")
    print("\t5 - Prefer not to say")
    gender = input("Enter here: ")
    time.sleep(0.5)

    # Stores user's age as the variable 'age'
    print("\nWhat is your age? Please enter it as a number below.")
    age = input("Enter here: ")
    time.sleep(0.5)

    # Stores user's sleep as variable 'sleep'
    print("\nHow many hours did you sleep last night? Please enter it as a number below.")
    sleep = input("Enter here: ")
    time.sleep(0.5)

    # Stores user's caffeine consumption as variable 'caffeine'
    print("\nHave you consumed any coffee or energy drinks today? Please enter 'Y' for yes, 'N' for no.")
    caffeine = input("Enter here: ")
    caffeine = caffeine.title()
    time.sleep(1)
    clear_output(wait=False)

    print(message_1)
    time.sleep(1.5)

    print(message_2)
    time.sleep(2.5)

    print(message_3)
    time.sleep(2.6)

    print(message_4)
    time.sleep(2.3)

    print(message_5)
    time.sleep(1.85)
    clear_output(wait=False)

    # Countdown to start of test
    for i in range(5):
        count = 5 - i
        print(count)
        time.sleep(1)
        clear_output(wait=False)

    print(message_6)
    time.sleep(1)
    clear_output(wait=False)

    return username, gender, age, sleep, caffeine
```

```python
# Function calls on specific images
# for questions requiring visual aid

def test_image(i):
    
    # Displaying image for question 2
    if i == 1:
        display(image1_q2)

    # Displaying image for question 4
    elif i == 3:
        display(image1_q4)

    # Displaying image for question 8
    elif i == 7:
        display(image2_q3)
        
    # Displaying image for question 9
    elif i == 8:
        display(image2_q4)
```

```python
# Function outputs the questions, 
# tests user's response against the answers, 
# and adds points for each correct response

def crunch(i):
    question = question_list[i]
    answer = answer_list[i]
    user_input = ""
    attempts = 2
    point = 0

    # Displays image 1 for first set of questions
    if i == 0:
        display(image1)
        time.sleep(10)
        clear_output(wait=False)

    # Displays image 2 for second set of questions
    elif i == 5:
        display(image2)
        time.sleep(10)
        clear_output(wait=False)

    while True:
        # Breaks loop (proceeds to next question) 
        # when all attempts have been used
        if attempts == 0:
            break

        # Displays images for specified questions
        test_image(i)
        
        print(f"Q{i+1}. {question}")
        user_input = input("Answer: ")
        user_input = user_input.lower()

        # Breaks loop (proceeds to next question) 
        # and adds 1 point 
        # when user response is correct
        if user_input == answer:
            point = point + 1
            clear_output(wait=False)
            break
            
        attempts = attempts - 1
        clear_output(wait=False)

    return point
```

```python
# Function prints closing messages

def print_closing():
    print(message_7)
    time.sleep(1)
    
    print(message_8)
    time.sleep(1.5)
```

```python
# Main function
# Collates all points collected by crunch() 
# and returns all data to be saved for storage

def memory_test():

    userpoints = 0

    # Runs introduction while collating all data 
    # from print_introduction()
    username, gender, age, sleep, caffeine = print_introduction()

    # Runs crunch function for each iteration 
    # of loop while collating points from crunch()
    for i in range(0,10):
        point = crunch(i)

        userpoints = userpoints + point

    print_closing()

    print(f"You scored {userpoints} out of 10 points.")
    time.sleep(2)

    # Return values to be stored as data
    return username, gender, age, sleep, caffeine, userpoints
```

```python
# Function confirming whether to save data

def confirm_save():
    confirm = ""
    print("Would you like to save your result? Type 'Yes' or 'No'")
    confirm = input("")
    confirm = confirm.title()
    if confirm == "Yes" :
        return True
    return False
```

```python
# Saving data to Google form

import requests
from bs4 import BeautifulSoup
import json 

# Google form URL
form_url = 'https://docs.google.com/forms/d/e/1FAIpQLSdFXbkH_hladdGFAQJLhopGLSHgvfge2BbbqO_jtdHf5bfwRA/viewform?vc=0&c=0&w=1&flr=0'

def send_to_google_form(data_dict, form_url):
    ''' Helper function to upload information to a corresponding google form 
        You are not expected to follow the code within this function!
    '''
    form_id = form_url[34:90]
    view_form_url = f'https://docs.google.com/forms/d/e/{form_id}/viewform'
    post_form_url = f'https://docs.google.com/forms/d/e/{form_id}/formResponse'

    page = requests.get(view_form_url)
    content = BeautifulSoup(page.content, "html.parser").find('script', type='text/javascript')
    content = content.text[27:-1]
    result = json.loads(content)[1][1]
    form_dict = {}
    
    loaded_all = True
    for item in result:
        if item[1] not in data_dict:
            print(f"Form item {item[1]} not found. Data not uploaded.")
            loaded_all = False
            return False
        form_dict[f'entry.{item[4][0][0]}'] = data_dict[item[1]]
    
    post_result = requests.post(post_form_url, data=form_dict)
    return post_result.ok
```

```python
# Run Memory test
username, gender, age, sleep, caffeine, userpoints = memory_test()

# Saves and uploads data to Google Form 
if(confirm_save()):
    saved_names.append(username)
    saved_genders.append(gender)
    saved_ages.append(age)
    saved_sleep.append(sleep)
    saved_caffeine.append(caffeine)
    saved_scores.append(userpoints)
    send_to_google_form(saved_data, form_url)
```
